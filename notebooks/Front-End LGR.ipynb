```python
import streamlit as st
import pandas as pd
import requests
from PIL import Image
from io import BytesIO

# Specify the path to the CSV file containing game data
file_path = "/Users/benjaminmonin/Documents/Legendary game recs/all_game_data_v1_corrected2"

# Read the CSV file into a DataFrame
df = pd.read_csv(file_path)

st.title("Game Recommender")

# Create a search input for filtering games
search_query = st.text_input("Type the title of a game:")

# Display the filtered game titles based on the search query
if search_query:
    filtered_games = df[df['title'].str.contains(search_query, case=False)]
    if not filtered_games.empty:
        st.write("Recommendations:")
        selected_game = st.selectbox("Select a game:", filtered_games['title'])
        
        # Get the selected game's URL and image URL
        selected_game_row = filtered_games[filtered_games['title'] == selected_game]
        game_url = selected_game_row['url'].values[0]
        image_url = selected_game_row['image'].values[0]
        
        # Display the selected game's image
        if image_url:
            response = requests.get(image_url)
            if response.status_code == 200:
                img = Image.open(BytesIO(response.content))
                st.image(img, use_column_width=True, caption=f"Image for {selected_game}")

                # Button to download the image
                if st.button("Download Image"):
                    # Determine the file name from the game name
                    file_name = f"{selected_game.replace(' ', '_')}.jpg"

                    # Save the image to a file
                    with open(file_name, "wb") as file:
                        file.write(response.content)

                    # Provide a download link for the image
                    st.write(f"Download your image: [Download Image]({file_name})")
            else:
                st.write("Image not found.")
        else:
            st.write("No image URL available for this game.")
    else:
        st.write("No matching games found.")
```
