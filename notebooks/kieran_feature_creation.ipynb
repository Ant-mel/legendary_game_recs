```python
df.loc[df['release_date'] < '1983-01-01', 'generation'] = '1'
```

```python
df.loc[(df['release_date'] >= '1983-01-01') & (df['release_date'] < '1995-01-01'), 'generation'] = '2'
```

```python
df.loc[(df['release_date'] >= '1995-01-01') & (df['release_date'] < '2006-01-01'), 'generation'] = '3'
```
